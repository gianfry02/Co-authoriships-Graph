Creazione grafo con metodo preferential attachment corretto che prende in considerazione anche il grado dei nodi aumentandoli ogni volta che aggiungi un arco

```python

import random

# Leggi il grafo esistente
with open('graphs/coauthorship_graph.pickle', 'rb') as f:
    G = pickle.load(f)

# Genera un nuovo grafo con gli stessi nodi
G_Preferential_Attachment_correct = G.copy()

# Numero di archi da aggiungere usando il Preferential Attachment
num_edges_to_add = 50

# Calcolo i punteggi di Preferential Attachment per tutte le coppie di nodi
pa_scores = [(node1, node2, score) for node1, node2, score in nx.preferential_attachment(G_Preferential_Attachment_correct)]

# Calcolo il grado di ogni nodo nel grafo
degree_dict = dict(G_Preferential_Attachment_correct.degree())

# Calcolo la somma dei gradi dei nodi nel grafo
total_degree = sum(degree_dict.values())

# Aggiungi archi in base ai punteggi di Preferential Attachment considerando il grado dei nodi
while G_Preferential_Attachment_correct.number_of_edges() < 161:
    # Calcolo la probabilità di connessione tra i nodi basata sul Preferential Attachment e sul grado
    probabilities = [(node1, node2, score * (degree_dict[node1] + degree_dict[node2]) / (2 * total_degree))
                     for node1, node2, score in pa_scores]
    
    # Scelgo casualmente un'arco in base alle probabilità calcolate
    chosen_edge = random.choices(probabilities, weights=[p[2] for p in probabilities])[0]

    if(chosen_edge[0] != chosen_edge[1]):
        # Aggiungo l'arco al grafo
        G_Preferential_Attachment_correct.add_edge(chosen_edge[0], chosen_edge[1])

        degree_dict[chosen_edge[0]] =+ 1
        degree_dict[chosen_edge[1]] =+ 1
        
        total_degree +=2

    

# Salva il nuovo grafo
with open('graphs/extended_coauthorship_graph_correct.pickle', 'wb') as f:
    pickle.dump(G_Preferential_Attachment_correct, f, pickle.HIGHEST_PROTOCOL)

# Leggi il nuovo grafo
with open('graphs/extended_coauthorship_graph_correct.pickle', 'rb') as f:
    G_Preferential_Attachment_correct = pickle.load(f)

# Visualizzazione del nuovo grafo
node_colors_new = []
node_labels_new = {}
for node, degree in G_Preferential_Attachment_correct.degree():
    if degree == 1:
        node_colors_new.append('gray')
    elif 2 <= degree <= 10:
        node_colors_new.append('blue')
    elif 11 <= degree <= 20:
        node_colors_new.append('purple')
    else:
        node_colors_new.append('yellow')

    node_labels_new[node] = node

plt.figure(figsize=(15, 15))
nx.draw_networkx(G_Preferential_Attachment_correct, pos=nx.spring_layout(G_Preferential_Attachment_correct), with_labels=True, node_color=node_colors_new,
                 width=1, font_size=10, node_size=100)

# Salva la visualizzazione del nuovo grafo
plt.savefig('visualizations/extended_graph_correct.pdf')
plt.show()
```

g. Riassumere le informazioni in un DataFrame, dove ogni riga rappresenta le informazioni relative ad un grafo, ed ogni colonna le informazioni relative ad una misura calcolata per quel grafo. Salvare nella cartella /results tale DataFrame.


```python
# Leggi i grafi
with open('graphs/coauthorship_graph.pickle', 'rb') as f:
    G = pickle.load(f)

with open('graphs/extended_coauthorship_graph.pickle', 'rb') as f:
    G_Prefential_Attachment = pickle.load(f)

with open('graphs/extended_coauthorship_graph_correct.pickle', 'rb') as f:
    G_Preferential_attachment_correct = pickle.load(f)

def calcola_misure(grafo, nome):
    clustering_medio = nx.average_clustering(grafo)
    centro = nx.center(grafo)
    
    # Converti la lista  in una stringa separata da virgole, se necessario
    centro_str = ', '.join(centro)

    raggio = nx.radius(grafo)
    distanza_media = nx.average_shortest_path_length(grafo)
    transitivita = nx.transitivity(grafo)
    omega = nx.omega(grafo)
    sigma = nx.sigma(grafo)

    risultati_misure = {
        "Nome Grafo": nome,
        "Clustering Medio": clustering_medio,
        "Centro del Grafo": centro_str,
        "Raggio": raggio,
        "Distanza Media": distanza_media,
        "Transitività": transitivita, 
        #confrontare questi valori tra il grafo originale e il nuovo grafo per ottenere un'indicazione della presenza di proprietà "small-world". 
        "Omega": omega, #omega vicino a zero indica una bassa lunghezza cammino breve medio
        "Sigma": sigma  #sigma maggiore 1 forte clusterizzazione
    }
    return risultati_misure


#Calcola le misure per entrambi i grafi
risultati_grafo_normale = calcola_misure(G, "Grafo normale")
risultati_grafo_preferential_attachment = calcola_misure(G_Prefential_Attachment, "Grafo Preferential Attachment")
risultati_grafo_preferential_attachment_corretto = calcola_misure(G_Preferential_attachment_correct, "G_Preferential_attachment_correct")

# Crea un DataFrame
Graphs_information_DF = pd.DataFrame([risultati_grafo_normale, risultati_grafo_preferential_attachment, risultati_grafo_preferential_attachment_corretto])

# Salva il DataFrame
Graphs_information_DF.to_csv('results/overall measures.csv', index=False)

# Visualizza il DataFrame
display(Graphs_information_DF)

#COME POSSIAMO NOTARE IN ENTRAMBI I CASI DEL GRAFO CON L' IMPLEMENTAZIONE DEL PREFERENTIAL ATTACHMENT SIGMA è MAGGGIORE A 1, QUESTO FA SI CHE VENGANO CLASSIFICATI COME SMALL-WORLD. I PUO NOTARE CHE COMUNQUE L'OMEGA è RELATIVAMENTE BASSO QUINDI IN ENTRAMBI I PERCORSI TRA I NODI SONO RELATIVAMENTE CORTI
```

```python
import scipy as sp

# Leggi i grafi
with open('graphs/coauthorship_graph.pickle', 'rb') as f:
    G = pickle.load(f)

with open('graphs/extended_coauthorship_graph.pickle', 'rb') as f:
    G_Prefential_Attachment = pickle.load(f)

with open('graphs/extended_coauthorship_graph_correct.pickle', 'rb') as f:
    G_Preferential_attachment_correct = pickle.load(f)

# Funzione per calcolare le misure di centralità per ogni nodo
def calcola_centralita(grafo, nome_grafo):
    degree_centrality = nx.degree_centrality(grafo)
    betweenness_centrality = nx.betweenness_centrality(grafo)
    closeness_centrality = nx.closeness_centrality(grafo)
    pagerank = nx.pagerank(grafo)
    hits = nx.hits(grafo)

    #Estrai i valori di hubness e authority da HITS
    hubness = hits[0]
    authority = hits[1]

    # Crea un DataFrame
    df_centralita = pd.DataFrame({
        "Tipologia grafo": nome_grafo,
        "Node": list(grafo.nodes),
        "Degree Centrality": list(degree_centrality.values()),
        "Betweenness Centrality": list(betweenness_centrality.values()),
        "Closeness Centrality": list(closeness_centrality.values()),
        "Pagerank": list(pagerank.values()),
        "Hubness": list(hubness.values()),
        "Authority": list(authority.values())
        
    })

    return df_centralita


# Calcola le misure di centralità per entrambi i grafi
df_centralita_original = calcola_centralita(G, "Grafo Originale")
df_centralita_preferential_attachment = calcola_centralita(G_Prefential_Attachment, "Grafo preferential attachment")
df_centralita_preferential_attachment_correct = calcola_centralita(G_Preferential_Attachment_correct, "Grafo preferential attachment correct")

# i risultati in un tre DataFrame
df_centralita_original = pd.DataFrame(df_centralita_original)
df_centralita_preferential_attachment = pd.DataFrame(df_centralita_preferential_attachment)
df_centralita_preferential_attachment_correct= pd.DataFrame(df_centralita_preferential_attachment_correct)

# Salva il DataFrame
df_centralita_original.to_csv('results/first_graph_centralies.csv', index=False)
df_centralita_preferential_attachment.to_csv('results/extended_graph_centralies.csv', index=False)
df_centralita_preferential_attachment_correct.to_csv('results/extended_graph_centralies_correct.csv', index=False)

# Visualizza il DataFrame
display(df_centralita_original)
display(df_centralita_preferential_attachment)
display(df_centralita_preferential_attachment_correct)
```

```python
from pyvis.network import Network

# Leggi i grafi
with open('graphs/coauthorship_graph.pickle', 'rb') as f:
    G = pickle.load(f)

with open('graphs/extended_coauthorship_graph.pickle', 'rb') as f:
    G_Prefential_attachment = pickle.load(f)

with open('graphs/extended_coauthorship_graph_correct.pickle', 'rb') as f:
    G_Preferential_attachment_correct = pickle.load(f)

#funzione per visualizzare interattivamente i grafi
def visualizza_grafo_interattivo(grafo, nome_grafo):
    nt = Network(
        height = "1300px",
        width= "100%"
    )
    
    #aggiungo i nodi
    for node, data in grafo.nodes(data=True):
        
         # Aggiungo nodi con colorazione in base al grado
        degree = grafo.degree(node)
        color = None
        if degree == 1:
            color = 'gray'
        elif 2 <= degree <= 10:
            color = 'blue'
        elif 11 <= degree <= 20:
            color = 'purple'
        else:
            color = 'yellow'
        
        nt.add_node(node, label=node, title=f"Affiliations: {data.get('affiliations', '')}\nCited by: {data.get('cited_by', '')}\nInterests: {data.get('interests', '')}", color=color)

        #andando nella pagina html si può notare che scorrendo il mouse sopra i nodi compaiono le etichette con le informazioni riguardanti a quel nodo
                                                
    #aggiungo gli archi
    for edge in grafo.edges():
        nt.add_edge(edge[0], edge[1])
    
    # Imposta stili e disposizione per una visualizzazione più ordinata
    nt.set_options("""
    var options = {
    "nodes": {
    "shape": "dot",
    "size": 12,
    "font": {
      "size": 12,
      "color": "black"
    },
    "borderWidth": 2,
    "color": {
      "border": "black"
    }
    },
    "edges": {
    "color": "gray",
    "width": 0.5,
    "smooth": {
      "enabled": true,
      "type": "continuous"
    },
    "arrows": {
      "to": {
        "enabled": false,
        "type": "arrow",
        "scaleFactor": 0.5
      }
    }
    },
    "physics": {
    "barnesHut": {
      "gravitationalConstant": -10000,
      "centralGravity": 0.3,
      "springLength": 150
    },
    "minVelocity": 0.75,
    "springConstant": 1000
    },
    "layout": {
    "randomSeed": 42,
    "improvedLayout": true,
    "hierarchical": {
      "enabled": false,
      "direction": "UD",
      "sortMethod": "directed"
    }
  }
}
""")

    nt.show(f"visualizations/{nome_grafo}.html", notebook=False,)

visualizza_grafo_interattivo(G, 'interactive_first_graph')
visualizza_grafo_interattivo(G_Prefential_attachment, 'interactive_extended_graph')
visualizza_grafo_interattivo(G_Preferential_attachment_correct, 'interactive_extended_correct_graph')

```
